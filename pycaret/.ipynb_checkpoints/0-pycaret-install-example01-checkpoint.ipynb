# Ejemplo simple de clasificación binaria

- Este ejemplo fue tomado del [repositorio de Github de pycaret - tutoriales.](https://github.com/pycaret/pycaret/blob/master/tutorials/Binary%20Classification%20Tutorial%20Level%20Beginner%20-%20%20CLF101.ipynb)

## Datos (créditos)

Estos datos están incorporados dentro del módulo pycaret.

```python
from pycaret.datasets import get_data
dataset = get_data('credit')
```

```python
# Dimensiones de la base de datos
dataset.shape
```

```python
# nombre de columnas: en pycaret es necesario que nuestras variables (columnas) tengan el atributo nombre
dataset.columns
```

# Train - Test

```python
data = dataset.sample(frac=0.95, random_state=786)
data_unseen = dataset.drop(data.index)
data.reset_index(inplace=True, drop=True)
data_unseen.reset_index(inplace=True, drop=True)
print('Datos para modelación (train): ' + str(data.shape))
print('Datos para predicción (test): ' + str(data_unseen.shape))
```

# Etapa 1: configurar entorno de pycaret

- La configuración del entorno se lleva a cabo con la función `setup()`. Esta función requiere dos argumentos obligatorios:
    - `data`: base de datos con la cual se ajustan los modelos
    - `target`: variable respuesta
- Es necesario que las variables estén tipificadas de la manera correcta, ya que *pycaret* realiza algunos procedimientos de preprocesamiento internamente.
- Al ejecutar `setup()` con los argumentos necesarios, mostrará el tipo de cada variable, de ser correcta la tipificación se presiona *enter* y obtendremos una descripción detallada de los *inputs* y de los procedimientos de preprocesamiento aplicados, por ejemplo, el método de imputación es simple (por defecto). Este proceso es automático, *pycaret* está diseñado para *hacer mucho con poco código*.
- El argumento `session_id` actúa como *semilla* para garantizar que los resultados sean replicables.
- Internamente la función `setup()` asigna aproximadamente el 70% de los datos para *train* y el 30% restante lo usa para *test*.
- El número de variables en la data original (*Original Data*) ha aumentado por la codificación de variables categóricas. Inicialmente la base de dates contiene 24 columnas y después de aplicar `setup()` obtenemos 88 columnas.
- Por defecto se ajusta como método de validación la *validación cruzada estratificada* con $k = 10$ (*Fold Number*).

```python
from pycaret.classification import *
exp_clf101 = setup(data = data, target = 'default', session_id=123)
```

# Comparando modelos

- La [documentación de modelos de clasificación](https://pycaret.readthedocs.io/en/latest/api/classification.html) puede ser consultada para conocer todos los argumentos disponibles en el `setup()` de este tipo de modelos.
- De forma automática *pycaret* probará más de 15 modelos disponibles para clasificación binaria, además, muestra métricas de desempeño (error) para evaluar cuál es el mejor algoritmo.
- La última columna del resultado de ejecutar la función `compare_models()` es el tiempo de ejecución en segundos.
- Aunque se puede ajsutar de manera individual cada algoritmo, siempre será una buena aproximación tener resultados para varios modelos de forma simultánea.
- La tabla está ordenada de mayor a menor precisión (Accuracy), aunque se puede cambiar con el argumento `sort` dentro de `compare_models()`.
- Al terminar señalará en color amarillo los mejores resultados para cada métrica de error en cada uno de los modelos. Esto es importante para determinar tentativamente qué modelos podrían ser apilados (*stacking*) o ensamblados.

```python
best_model = compare_models()
```

```python
# El mejor modelo
best_model
```

# Ajustando Random Forest

- El ajuste de un modelo individual se puede obtener con la función [`create_model()`](https://pycaret.org/create-model/).
- Por defecto la función `create_model()` utiliza *validación cruzada estratificada* con $k = 10$. Este número puede ser alterado con el argumento `fold`.
- La función arroja una tabla resumen para diferentes métricas de error. Por defecto está organizada de mayor a menor precisión. Elegir la métrica de error adecuada es una característica importante cuando estamos ajustando modelos de machine learning, en [este blog](https://medium.com/@MohammedS/performance-metrics-for-classification-problems-in-machine-learning-part-i-b085d432082b) hay más información de cómo elegir la métrica idónea para problemas de clasificación.
- La tabla proporciona resultados de ajuste para cada pliegue de la validación cruzada, en este caso 10. Al final presenta en color amarillo el promedio de cada métrica. El resultado coincide con lo obtenido a través de la función `compare_models()`, esto sucede porque en ambos se entrenó con $k = 10$, además, en el `setup()` está declarada la *semilla* a través del argumento `session_id = 123`.

```python
rf = create_model('rf')
```

# Tuning de Random Forest

- Con la función `create_model()` se obtiene el ajuste con los hiperparámetros por defecto (prederminados), sin embargo, podemos ajustarlos a través de la función `tune_model()`.
- La función `tune_model()` evalúa los hiperparámetros a través de *búsqueda de cuadrícula aleatoria (Random Grid Search)* en un espacio de posibilidades predefinido. Es posible personalizar el espacio de búsqueda (cuadrícula) con el agurmento `custom_grid`.
- Por defecto el ajuste de hiperparámetros se hace optimizando la métrica *Accuracy*, sin embargo, puede ser editada a través del argumento `optimize`.
- Para mayor información de la función `tune_model()` consultar [la documentación.](https://pycaret.org/tune-model/)

```python
tuned_rf = tune_model(rf)
```

# Finalizando ajuste de modelo

- Todo proceso de modelamiento típico podría ser resumido como se muestra en la siguiente imagen (tomada del libro [Tidy Modeling With R - Max Kuhn y Julia Silge](https://www.tmwr.org/))

![](https://www.tmwr.org/premade/modeling-process.svg)

- Inicialmente exploramos nuestros datos, se incorpora alguna estrategia de ingeniería de características, se comparan modelos, evaluamos el rendimiento de los mismos y, posiblemente surgen algunos candidatos sobre los cuales haremos un proceso más exhaustivo de ajuste. Luego, terminamos de afinar nuestros modelos, usualmente con mayor rigor en la ingeniería de características, volvemos a evaluar nuestros algoritmos y, para terminar, finalizamos nuestro flujo de trabajo desplegando el o los mejores modelos. Este proceso lo podríamos resumir en las siguientes etapas, haciendo énfasis en que para cada una de ellas *pycaret* suministra alguna función que reduce tiempo y líneas de código:
    0. Depuración y estructuración de datos ⟶ Esta es una fase obligatoria del modelado, aunque no es inherente al ajuste de modelos, por tal razón está fuera del alcance de *pycaret* y no se muestra en el esquema anterior. Bibliotecas como *pandas* en Python y *dplyr* en R, son herramientas flexibles y de gran utilidad para dar forma a nuestros datos.
    1. Análisis exploratorio de datos ⟶ Las herramientas de visualización como *ggplot2* de R, *plotly* (R y Python) o *matplotlib* (Python), son buenas opciones para explorar nuestros datos.
    2. Preprocesamiento e ingeniería de características ⟶ La función `setup()` permite de manera fácil el preprocesamiento y algunas estrategias de ingeniería de características.
    3. Comparación de modelos ⟶ La función `compare_models()` es una excelente herramienta que permite evaluar rapidamente cuáles modelos tienden a comportarse mejor para nuestro problema de predicción. En esta fase también podríamos utilizar la función `create_model()`.
    4. Refinación de modelos ⟶ La función `tune_model()` permite *afinar" nuestros modelos en aras de encontrar la combinación de hiperparámetros óptima.
    5. Modelo final ⟶ la función `finalize_model()` permite ajustar el modelo con todos los datos, incluido el de validación o test (interno). Con este modelo final podemos realizar predicciones sobre observaciones no vistas por el modelo.

```python
final_rf = finalize_model(tuned_rf)
```

```python
final_rf
```

# Predicción en nuevos datos

```python
unseen_predictions = predict_model(final_rf, data=data_unseen)
unseen_predictions.head()
```

# Exportando el modelo final

- La función `save_model()` permite exportar un modelo para que pueda ser utilizado posteriormente con nuevos datos.
- Para mayor información consultar [la documentación.](https://pycaret.org/save-model/)
- El modelo se guarda con extensión *.pkl*

```python
save_model(final_rf, 'modeloRF')
```
